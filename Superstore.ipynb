```python
# 8. Feature Engineering:
#   1. Create new columns:
#     1. Original Price: The price before any discount is applied.
    # 2. Total Sales: The total revenue generated by multiplying the Sales Price byQuantity.
    # 3. Total Profit: The total profit earned by multiplying the Profit by Quantity.
    # 4. Discount Price: The amount of discount applied, calculated based on theOriginal Price and Discount.
    # 5. Total Discount: The total discount value for the quantity sold.

df['Original Price'] = df['Sales Price'] / (1 - df['Discount'])
df['Total Sales'] = df['Sales Price'] * df['Quantity']
df['Total Profit'] = df['Profit'] * df['Quantity']
df['Discount Price'] = df['Original Price'] - df['Sales Price']
df['Total Discount'] = (df['Original Price'] * df['Discount']) * df['Quantity']
df[['Original Price', 'Sales Price', 'Quantity', 'Total Discount']].head()
```

```python
# 2. Create a function called remove_outliers which takes as parameter the dataframe
# and the column which needs to be searched for outliers. Using the 3 * /QR rule, the
# function should detect and remove the outliers to return the cleaned dataframe

# Why 3*/QR?
# The 3 *!QR method is applied in situations where the dataset has a high variance, and

# the standard 1.5 * QR might flag too many points as outliers. This method ensures that
# only the most extreme values are removed, preserving the integrity of the dataset while
# still mitigating the influence of true outliers


def remove_outliers(df, column):
    # Step 1: Calculate Q1, Q3, IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Step 2: Define bounds using 3 * IQR
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    # Step 3: Filter dataframe
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Step 4: Print summary (optional but good for debugging)
    print(f"Column: {column}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")

    return df_clean
```

```python
df = remove_outliers(df, 'Sales Price')
```

```python
def remove_outliers_multi(df, columns):
    for col in columns:
        df = remove_outliers(df, col)
    return df
```

```python
# 10. Customer Segmentation and Analysis:
#     1. Calculate Customer Sales Quintile and Customer Profit Quintile based on total sales and total profit per Customer ID.
    # 2. What is a Quintile? Quintiles are a statistical way of dividing data into five equal parts, each representing 20% of the data.
       # For example, customers in the top quintile (Q5) represent the top 20% of sales or profit.

customer_df = df.groupby('Customer ID').agg({
    'Total Sales': 'sum',
    'Total Profit': 'sum'
}).reset_index()

customer_df['Sales_Quintile'] = pd.qcut(
    customer_df['Total Sales'], 
    5, 
    labels=['Q1','Q2','Q3','Q4','Q5'],
    duplicates='drop'
)

customer_df['Profit_Quintile'] = pd.qcut(
    customer_df['Total Profit'], 
    5, 
    labels=['Q1','Q2','Q3','Q4','Q5'],
    duplicates='drop'
)
```

```python
# 6. Temporal Analysis:
# 1. Sales and Profit Trends Over Time: Use a time series plot to analyze how sales and profit have trended over the years or months. 
#     This will help in identifying any seasonal patterns.
# Create Year-Month column
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Aggregate
trend = df.groupby('YearMonth')[['Total Sales', 'Total Profit']].sum().reset_index()

# Convert to string for plotting
trend['YearMonth'] = trend['YearMonth'].astype(str)

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(trend['YearMonth'], trend['Total Sales'], label='Sales')
plt.plot(trend['YearMonth'], trend['Total Profit'], label='Profit')

plt.xticks(rotation=45)
plt.title('Sales and Profit Trend Over Time')

plt.legend()
plt.show()
```

```python
# 2. Order Frequency by Month: Use a bar chart or line plot to show the number of orders placed each month. Highlight any months with
#     unusually high or low order frequencies.
df['Month'] = df['Order Date'].dt.month_name()

order_freq = df.groupby('Month')['Order ID'].count()

# Sort months properly
order_freq = order_freq.reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])

order_freq.plot(kind='bar', figsize=(10,5))

plt.title('Order Frequency by Month')
plt.ylabel('Number of Orders')

plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Yearly Growth in Sales and Profit: Use a year-over-year growth chart to compare the sales and profit growth over different years.
# Extract Year
df['Year'] = df['Order Date'].dt.year

yearly = df.groupby('Year')[['Total Sales', 'Total Profit']].sum()

# Growth calculation
growth = yearly.pct_change() * 100

print(growth)

growth.plot(kind='bar', figsize=(8,5))

plt.title('Year-over-Year Growth (%)')
plt.ylabel('Growth %')

plt.show()
```
